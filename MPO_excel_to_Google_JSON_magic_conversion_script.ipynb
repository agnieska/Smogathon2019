```python
column_weeks =[]
column_weekdays = []
i = 0
for line in l :
    line = line.strip(" ")
    i = i + 1
    print(i)
    print(line)
    #print(type(line))
    weeks = set()
    days = set()
    if line == "1x / dzień":
        weeks = {1,2,3,4,5}
        days = {"pn", "wt", "śr", "czw", "pt", "sb", "nd"}
    else :
        n = line.split("-")
        #print(n)
        for el in n :
            week = el.split(".")[0].strip(" ").lower()
            if week == 'k':
                weeks = {1,2,3,4,5}
            else :
                weeks.add(int(week))
            day = el.split(".")[1].strip(" ").lower()
            days.add(day)    
    #days = list(set(days))
    column_weeks.append(weeks)
    column_weekdays.append(days)
        
```

```python
df["weekdays"] = column_weekdays
df["weeks"]= column_weeks
```

```python
i = 0
for i in cleaning_dict :
    #print("i=", item)
    weeks = cleaning_dict[i]['weeks']
    #print("weeks",weeks)
    for week in weeks :    
        #print("week=",week)
        week = int(week)
        #print("week=",week)
        for weekday in cleaning_dict[i]['weekdays']:
            #print("weekday=",weekday)
            for z in calendar_dict:
                calendar_item = calendar_dict[z]
                #print("cal=", calendar_item)
                date = calendar_item['date'].strip('"')
                if calendar_item[weekday] == float(week):
                    street = cleaning_dict[i]['street']
                    #print(date)
                    #print(street)
                    result_dict[date].add(street)
                    #print(result_dict[date])
    i = i + 1
```

```python
def save_json(my_dict, filename):
    with open (filename, "w", encoding="utf-8") as file: #"iso8859-1"
        json.dump(my_dict, file, indent=2, allow_nan=False, ensure_ascii=False, sort_keys = True)
```

```python
for date in dates:
    d = list(result_dict[date])
    filename = date.replace("/","-").replace(".", "-")+".json"
    save_json(d, filename)
```

```python
for date in dates: 
    d = list(result_dict[date])
    df = pd.DataFrame(d)
    date = date.replace("/","-").replace(".", "-")
    df.columns = [date]
    filename = date +".csv"
    df.to_csv(filename, header=False, index=False, encoding="utf-8")
```
